```python
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
```

```python
sns.set_style("darkgrid")
matplotlib.rcParams["font.size"] = 14
matplotlib.rcParams["figure.figsize"] = (9,5)
matplotlib.rcParams["figure.facecolor"] = "#00000000"
```

<h3>Analasis exploratorio y visualizacion</h3>

```python
raw_data = [
    'Coarse wool Price',
 'Copra Price',
 'Cotton Price',
 'Fine wool Price',
 'Hard log Price',
 'Hard sawnwood Price',
 'Hide Price',
 'Plywood Price',
 'Rubber Price',
 'Softlog Price',
 'Soft sawnwood Price',
 'Wood pulp Price',
]
```

```python
corrmat = data[raw_data].corr()
```

```python
corrmat
```

```python
fig = plt.figure(figsize=(12,9))
mask = np.triu(np.ones_like(corrmat,dtype=bool))
sns.heatmap(corrmat,vmax=.8,mask=mask,square=True,annot=True)
```

```python
plt.figure(figsize=(15,10))
plt.plot(data[raw_data])
plt.title("Comparacion de precios - materias primas")
plt.xlabel("Years")
plt.ylabel("Precios")
plt.legend(raw_data)
```
